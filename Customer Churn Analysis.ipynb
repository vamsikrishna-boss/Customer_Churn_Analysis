```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

## Calculate the Percentages
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack()*100

## Plot
fig, ax = plt.subplots(figsize=(4, 4))

## plot the bars
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])

## Add percentage lables on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width /2, y + height /2, f'{height:.1f}%', ha='center', va='center')

plt.title('Churn by senior Citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('parcentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', loc='upper right')

plt.show()

```

```python
#Comparative a greater parcentage of people in senior citizen category have churned
```

```python
# List of columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the subplot grid
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols
plt.figure(figsize=(18, 12))

for i, col in enumerate(columns, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.countplot(data=df, x=col, hue="Churn")
    plt.title(f"{col} vs Churn")
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.suptitle("Service Usage vs Churn", fontsize=18, y=1.02)
plt.show()
```

## Type : 2

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the subplot grid
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols

# Create Subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Iterate over columns and plot count plots
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i])
    axes[i].set_title(f'{col} vs Churn')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

# Remove empty subplots if columns < total subplots
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.suptitle("Service Usage Features vs Churn", fontsize=20, y=1.02)
plt.show()
```

```python
# The visual summary shows that customers who lack security or support services (e.g., no TechSupport, OnlineSecurity, or DeviceProtection) tend to churn more frequently. Fiber optic users have higher churn rates compared to DSL or no internet service. Additionally, those with active streaming and backup services also show noticeable churn, indicating that expectation gaps or service dissatisfaction could be influencing customer retention.
```
